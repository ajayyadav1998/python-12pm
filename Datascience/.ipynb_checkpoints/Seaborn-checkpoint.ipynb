```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
plt.rcParams
```

```python
sns.get_dataset_names()
```

```python
tip = sns.load_dataset('tips')
```

```python
tip.head()
```

```python
tip.info()
```

```python
color = tip['sex'].map({'Female':'red','Male':'green'})
```

```python
color
```

```python
tip.sex.value_counts().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tip.total_bill,tip.tip,color=color)
plt.text(15,5,s='Female',color='red',fontdict={'fontsize':15})
plt.text(15,6,s='Male',color='green',fontdict={'fontsize':15})
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#99ff99')
plt.rcParams['text.color'] = 'red'
sns.countplot(x='sex',data=tip,palette='cubehelix')
plt.title('plot')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#800080')
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tip)
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#800080')
sns.scatterplot(x='total_bill',y='tip',hue='sex',size='smoker',palette='BuPu_r',data=tip)
plt.show()
```

```python
sns.palplot(sns.color_palette('BuPu'),size=10)
sns.palplot(sns.color_palette('Dark2'),size=10)
sns.palplot(sns.color_palette('cubehelix'),size=10)
sns.palplot(sns.color_palette('Set1'),size=10)
```

```python
tip.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='sex',hue='smoker',data=tip,palette='tab20b_r')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='sex',hue='smoker',data=tip,palette='tab20b_r',dodge=False,alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='lightgreen')
sns.distplot(a=tip['total_bill'],hist_kws={'color':'green','edgecolor':'k'},bins=20,
            kde_kws={'color':'k','lw':4,'label':'kde'})
plt.show()
```
